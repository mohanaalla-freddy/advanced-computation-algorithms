```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
import numpy as np


data_path = "/Users/mohanavenkataphaneendrareddyalla/Downloads/DemoExample_GeneExpression.csv"

data = pd.read_csv(data_path, sep="\t")
print(data.head())
print(data.info())
data.set_index(data.columns[0], inplace=True)
if not all(data.applymap(lambda x: isinstance(x, (int, float)) or pd.isna(x)).all()):
    data = data.apply(pd.to_numeric, errors='coerce')  # Convert non-numeric to NaN
    data = data.fillna(data.mean())  # Fill NaNs with column means
# Optional: Standardize the data
data = (data - data.mean()) / data.std()

# Define metrics and methods
distance_metrics = ['euclidean', 'cosine', 'correlation']
linkage_methods = ['single', 'complete', 'average', 'weighted']
for metric in distance_metrics:
    for method in linkage_methods:
        try:
            # Generate clustermap
            clustermap = sns.clustermap(data, metric=metric, method=method, cmap="vlag")
            
            # Save figure with descriptive filename
            output_path = f'/Users/mohanavenkataphaneendrareddyalla/Desktop/Homework_6/clustermap_{metric}_{method}.png'
            clustermap.savefig(output_path)
            plt.close()
            print(f"Saved: {output_path}")
        except Exception as e:
            print(f"Error for metric: {metric}, method: {method} -> {e}")
```
