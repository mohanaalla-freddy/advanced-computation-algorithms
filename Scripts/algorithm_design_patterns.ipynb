```python
import pandas as pd
from scipy.sparse.csgraph import minimum_spanning_tree
import numpy as np

# Loading the dataset 
file_path = '/Users/mohanavenkataphaneendrareddyalla/Desktop/mutation_matrix.csv'
data = pd.read_csv(file_path)

root_node = "NC_045512"
distance_matrix = data.set_index("Unnamed: 0")


distances = distance_matrix.to_numpy()
mst = minimum_spanning_tree(distances).toarray()
edges = []
for i in range(mst.shape[0]):
    for j in range(mst.shape[1]):
        if mst[i, j] != 0:  
            edges.append((distance_matrix.index[i], distance_matrix.columns[j], mst[i, j]))

# Save the MST edges to a CSV file
mst_output_path = '/Users/mohanavenkataphaneendrareddyalla/Desktop/Adv_comp_asspects_in_bioinfo/Homework_5/minimum_spanning_tree.csv'
mst_df = pd.DataFrame(edges, columns=["Node1", "Node2", "Weight"])
mst_df.to_csv(mst_output_path, index=False)

print(f"Minimum spanning tree saved to {mst_output_path}")
```

```python
data = pd.read_csv(file_path)
root_node = "NC_045512"
distance_matrix = data.set_index("Unnamed: 0")


mutation_matrix = distance_matrix.to_numpy()
genomes = distance_matrix.columns

root_index = distance_matrix.columns.get_loc(root_node)

def maximum_flow_calculation(mutation_matrix, root_index, target_index):
    flow_value = mutation_matrix[root_index, target_index]  # Simplified for example
    return flow_value
flow_scores = {}
for i, genome in enumerate(genomes):
    if i == root_index:
        continue  # Skip the root node itself
    flow_value = maximum_flow_calculation(mutation_matrix, root_index, i)
    flow_scores[genome] = flow_value

ranking = sorted(flow_scores.items(), key=lambda x: x[1], reverse=True)
ranking_df = pd.DataFrame(ranking, columns=["Genome", "MFP_Score"])

ranking_output_path = '/Users/mohanavenkataphaneendrareddyalla/Desktop/Adv_comp_asspects_in_bioinfo/Homework_5/mfp_ranking.csv'
ranking_df.to_csv(ranking_output_path, index=False)

print(f"MFP ranking saved to {ranking_output_path}")
```
