```python
from Bio import SeqIO

input_file = "/Users/mohanavenkataphaneendrareddyalla/Downloads/sequence.fasta" 
output_file = "/Users/mohanavenkataphaneendrareddyalla/Desktop/Adv_comp_asspects_in_bioinfo/Homework_3/output_spike.fasta"  
# Open the output file to write the filtered sequences
with open(output_file, "w") as outfile:
    # Parse the input fasta file
    for record in SeqIO.parse(input_file, "fasta"):
        # Check if the protein name contains "SPIKE_"
        if "SPIKE_" in record.description:
            parts = record.description.split("|")
            if len(parts) > 1:
                # Get the ID from 'gi' or 'sp' and extract the protein symbol
                new_header = parts[1] + " " + parts[-1].split()[0]
                # Modifying the record's description with the new header
                record.description = new_header
                record.id = new_header  # Set the ID as well
                # Write the filtered record to the output file
                SeqIO.write(record, outfile, "fasta")
                print(f">{new_header}")
                print(record.seq
                     )
```
